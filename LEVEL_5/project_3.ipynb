```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\vatch\Downloads\titanic dataset.csv")
summary_stats = df[['Age', 'Fare']].describe()
print(" Summary Statistics:\n", summary_stats)
survival_rates = df.groupby('Pclass')['Survived'].mean().reset_index()
print("\n Survival Rates by Class:\n", survival_rates)
plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Survived', data=survival_rates, palette='coolwarm')
plt.title('Survival Rates by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate (%)')
plt.xticks([0, 1, 2], ['1st Class', '2nd Class', '3rd Class'])
plt.ylim(0, 1)
plt.show()
```
