```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("House Price India.csv")
summary_stats=df[['Price','Area of the basement']].describe()
df['Price_per_SqFt'] = df['Price'] / df['Area of the basement']
plt.scatter(df['Area of the basement'], df['Price'])
plt.title("House Price Explorer")
plt.xlabel("Area of the basement")
plt.ylabel("Price")
plt.show()
```
