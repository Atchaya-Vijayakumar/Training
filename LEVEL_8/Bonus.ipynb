```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.random.pareto(a=2,size=1000)
```

```python
plt.hist(data,bins=50,density=True,log=True)
plt.title("Pareto Distribution Analysis")
plt.xlabel("Value")
plt.ylabel("Log(Density)sorted_data = np.sort(data)[::-1]")
plt.show()
```

```python
sorted_data = np.sort(data)[::-1]
top_20_percent = sorted_data[:200]
sum_top_20 = top_20_percent.sum()
total_sum = sorted_data.sum()
```

```python
print(f"Sum of top 20%: {sum_top_20:.2f}")
print(f"Total sum: {total_sum:.2f}")
print(f"Percentage of total sum from top 20%: {100 * sum_top_20 / total_sum:.2f}%")
```
