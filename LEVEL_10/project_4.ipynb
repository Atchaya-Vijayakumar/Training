```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = "House Price India.csv" 
data = pd.read_csv(file_path)
data.head()
```

```python
correlation_value = data['Price'].corr(data['living area'], method='pearson')
print(correlation_value)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='living area', y='Price', data=data)
plt.title(f"Scatter Plot of Price vs. Living Area\nPearson Correlation: {correlation_value:.2f}")
plt.xlabel('Living Area (sq ft)')
plt.ylabel('Price (INR)')
plt.grid(True)
plt.show()
```
