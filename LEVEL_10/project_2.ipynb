```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('product_details.csv')
data.head()
data['Selling Price'] = pd.to_numeric(data['Selling Price'].str.replace('$', ''), errors='coerce')
data = data.dropna(subset=['Selling Price'])
price_range = np.ptp(data['Selling Price'])
price_variance = np.var(data['Selling Price'])
price_std_dev = np.std(data['Selling Price'])
price_stats = pd.DataFrame({
    'Metric': ['Range', 'Variance', 'Standard Deviation'],
    'Value': [price_range, price_variance, price_std_dev]
})
print(price_stats)
plt.figure(figsize=(8, 5))
sns.histplot(data['Selling Price'], bins=30, kde=True, color='skyblue')
plt.title('Price Distribution - Histogram')
plt.xlabel('Selling Price ($)')
plt.ylabel('Frequency')
plt.show()
plt.figure(figsize=(8, 5))
sns.boxplot(data['Selling Price'], color='lightcoral')
plt.title('Price Distribution - Boxplot')
plt.xlabel('Selling Price ($)')
plt.show()
```
